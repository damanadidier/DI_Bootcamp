# What is data analysis?
Data analysis is the process of inspecting, cleansing, transforming, and modeling data with the goal of discovering useful information, drawing conclusions, and supporting decision-making. It involves various techniques and tools to analyze raw data and extract meaningful insights that can inform business strategies, scientific research, and policy-making.
# Why is data analysis important in modern contexts?
Data analysis is crucial in modern contexts for several reasons. First, it helps organizations make informed decisions by providing evidence-based insights. In a world where data is generated at an unprecedented rate, the ability to analyze and interpret this data can give businesses a competitive edge. Second, data analysis can identify trends and patterns that may not be immediately apparent, allowing for proactive measures and strategic planning. Lastly, it can improve efficiency and reduce costs by optimizing processes and identifying areas for improvement.

1. **Healthcare**: Data analysis is used to improve patient outcomes, optimize treatment plans, and manage healthcare resources. By analyzing patient data, healthcare providers can identify trends in diseases, predict outbreaks, and personalize treatments based on individual patient profiles.

# Exercise 3: Identifying Data Types


# Objective: Learn to identify different data types.

For the datasets from the previous exercise, categorize each column of it as either quantitative or qualitative and explain your reasoning.

```python
import pandas as pd

# ── Load Dataset ───────────────────────────────────────────
sleep_df = pd.read_csv("sleep_data.csv")

# ── Step 1: Quick overview ─────────────────────────────────
print("=" * 55)
print("  DATASET: How Much Sleep Do Americans Really Get?")
print("=" * 55)
print(f"\n Shape: {sleep_df.shape[0]} rows × {sleep_df.shape[1]} columns")
print(f"\n Columns & Raw dtypes:\n{sleep_df.dtypes}")

# ── Step 2: Categorize each column ─────────────────────────
column_types = {
    "index": {
        "dtype"   : sleep_df["index"].dtype,
        "type"    : "Quantitative – Discrete",
        "reason"  : "Auto-generated row ID; whole numbers, no decimals"
    },
    "Year": {
        "dtype"   : sleep_df["Year"].dtype,
        "type"    : "Quantitative – Discrete",
        "reason"  : "Represents a calendar year (2003, 2004...); countable whole numbers"
    },
    "Period": {
        "dtype"   : sleep_df["Period"].dtype,
        "type"    : "Qualitative – Nominal",
        "reason"  : f"Text label (e.g. 'Annual'); no numeric meaning. Unique: {sleep_df['Period'].unique()}"
    },
    "Avg hrs per day sleeping": {
        "dtype"   : sleep_df["Avg hrs per day sleeping"].dtype,
        "type"    : "Quantitative – Continuous",
        "reason"  : "Decimal hours (e.g. 8.57); measurable on a continuous scale"
    },
    "Standard Error": {
        "dtype"   : sleep_df["Standard Error"].dtype,
        "type"    : "Quantitative – Continuous",
        "reason"  : "Statistical margin of error; decimal values on a continuous scale"
    },
    "Type of Days": {
        "dtype"   : sleep_df["Type of Days"].dtype,
        "type"    : "Qualitative – Nominal",
        "reason"  : f"Named day categories; no natural numeric order. Unique: {sleep_df['Type of Days'].unique()}"
    },
    "Age Group": {
        "dtype"   : sleep_df["Age Group"].dtype,
        "type"    : "Qualitative – Ordinal",
        "reason"  : "Age ranges with a clear order (15–24 < 25–34 < ... < 65+)"
    },
    "Activity": {
        "dtype"   : sleep_df["Activity"].dtype,
        "type"    : "Qualitative – Nominal",
        "reason"  : f"Type of activity label; no numeric meaning. Unique: {sleep_df['Activity'].unique()}"
    },
    "Sex": {
        "dtype"   : sleep_df["Sex"].dtype,
        "type"    : "Qualitative – Nominal",
        "reason"  : f"Gender categories; no natural order. Unique: {sleep_df['Sex'].unique()}"
    },
}

# ── Step 3: Display results ────────────────────────────────
print("\n\n DATA TYPE CATEGORIZATION")
print("=" * 55)
for col, info in column_types.items():
    print(f"\n🔹 Column     : {col}")
    print(f"   Raw dtype  : {info['dtype']}")
    print(f"   Data Type  : {info['type']}")
    print(f"   Reasoning  : {info['reason']}")

# ── Step 4: Summary table ──────────────────────────────────
summary = pd.DataFrame([
    {"Column": col, "Data Type": info["type"]}
    for col, info in column_types.items()
])

print("\n\n SUMMARY TABLE")
print("=" * 55)
print(summary.to_string(index=False))
```

# Exercise 4: Exploring Data Types


# Objective: Learn about different types of data in data analysis.
# Task:
Load the Iris dataset.
Identify and list which columns in your dataset are qualitative and which are quantitative.
Write a brief description of why each column is classified as qualitative or quantitative.

```python
import pandas as pd

# ── Step 1: Load the Dataset ───────────────────────────────
iris_df = pd.read_csv("Iris.csv")

# ── Step 2: Display first few rows ────────────────────────
print("=" * 55)
print("         DATASET: Iris Dataset")
print("=" * 55)
print(f"\n First 5 Rows:\n")
print(iris_df.head())
print(f"\n Shape: {iris_df.shape[0]} rows × {iris_df.shape[1]} columns")
print(f"\n Raw dtypes:\n{iris_df.dtypes}")

# ── Step 3: Classify each column ──────────────────────────
column_types = {
    "Id": {
        "type"   : "Quantitative – Discrete",
        "reason" : "Whole number row identifier (1, 2, 3...); "
                   "no decimals, but not meaningful for math"
    },
    "SepalLengthCm": {
        "type"   : "Quantitative – Continuous",
        "reason" : "Measured in cm with decimals (e.g. 5.1); "
                   "can take any value on a scale"
    },
    "SepalWidthCm": {
        "type"   : "Quantitative – Continuous",
        "reason" : "Measured in cm with decimals (e.g. 3.5); "
                   "continuous physical measurement"
    },
    "PetalLengthCm": {
        "type"   : "Quantitative – Continuous",
        "reason" : "Measured in cm with decimals (e.g. 1.4); "
                   "continuous physical measurement"
    },
    "PetalWidthCm": {
        "type"   : "Quantitative – Continuous",
        "reason" : "Measured in cm with decimals (e.g. 0.2); "
                   "continuous physical measurement"
    },
    "Species": {
        "type"   : "Qualitative – Nominal",
        "reason" : "Named flower categories (Iris-setosa, "
                   "Iris-versicolor, Iris-virginica); "
                   "no numeric value or natural order"
    },
}

# ── Step 4: Print results ──────────────────────────────────
print("\n\n COLUMN CLASSIFICATION")
print("=" * 55)
for col, info in column_types.items():
    print(f"\n🔹 Column    : {col}")
    print(f"   Type      : {info['type']}")
    print(f"   Reasoning : {info['reason']}")

# ── Step 5: Summary table ──────────────────────────────────
summary = pd.DataFrame([
    {"Column": col, "Classification": info["type"]}
    for col, info in column_types.items()
])

print("\n\n SUMMARY TABLE")
print("=" * 55)
print(summary.to_string(index=False))

# ── Step 6: Grouped view ───────────────────────────────────
print("\n\n QUANTITATIVE COLUMNS:")
quant = [c for c, i in column_types.items() if "Quantitative" in i["type"]]
print("  →", quant)

print("\n  QUALITATIVE COLUMNS:")
qual = [c for c, i in column_types.items() if "Qualitative" in i["type"]]
print("  →", qual)
```

Quantitative Columns
These columns contain numerical values that can be measured or counted.

1. Id — Quantitative (Discrete)
This column is a unique whole-number identifier assigned to each row (1, 2, 3… 150). It is quantitative because it holds numeric values, and discrete because it only takes integer values — there is no such thing as row 2.5. However, note that while technically numeric, Id carries no meaningful mathematical value (calculating the average Id would be meaningless).

2. SepalLengthCm — Quantitative (Continuous)
This column records the length of the sepal (the outer leaf of the flower) in centimeters. It is quantitative because it is a physical measurement expressed as a number, and continuous because it can take any decimal value within a range (e.g. 4.6, 5.1, 7.9). Operations like mean, min, and max are all meaningful here.

3. SepalWidthCm — Quantitative (Continuous)
This column records the width of the sepal in centimeters. Like SepalLengthCm, it is a continuous physical measurement with decimal precision (e.g. 3.0, 3.5). It supports all arithmetic operations and is commonly used to distinguish between species.

4. PetalLengthCm — Quantitative (Continuous)
This column records the length of the petal (the inner, often colorful part of the flower) in centimeters. It is continuous and quantitative for the same reasons as the sepal measurements. Petal length is one of the most informative features in this dataset for separating species.

5. PetalWidthCm — Quantitative (Continuous)
This column records the width of the petal in centimeters. It is a continuous quantitative variable with decimal values (e.g. 0.2, 1.8). Along with petal length, it is highly useful for classifying iris species and detecting patterns.

Qualitative Column
This column contains categories or labels, not numbers.

6. Species — Qualitative (Nominal)
This column identifies which of the three iris species each flower belongs to: Iris-setosa, Iris-versicolor, or Iris-virginica. It is qualitative because it represents a label, not a measurement. It is specifically nominal (not ordinal) because there is no natural ranking or order between the three species — none is "greater than" or "less than" another. This column is typically used as the target variable in classification tasks.

# Exercise 5: Basic Observation Skills in Data Analysis


# Objective: Develop observation skills for data analysis.
# Task:
Load the How Much Sleep Do Americans Really Get? dataset into Jupyter or Google Colab.
Identify columns that could be interesting for a specific type of analysis (e.g., trend analysis, group comparison) and explain your choice.

```python
import pandas as pd

# ── Step 1: Load Dataset ───────────────────────────────────
sleep_df = pd.read_csv("sleep_data.csv")

print("=" * 55)
print("   DATASET: How Much Sleep Do Americans Really Get?")
print("=" * 55)
print(f"\n First 5 Rows:\n{sleep_df.head()}")
print(f"\n Shape : {sleep_df.shape[0]} rows × {sleep_df.shape[1]} columns")
print(f"\n Year Range : {sleep_df['Year'].min()} – {sleep_df['Year'].max()}")

# ── Step 2: Identify Interesting Columns ──────────────────

# ANALYSIS 1 — Trend Analysis
print("\n\n TREND ANALYSIS: Sleep over time")
print("-" * 45)
trend = (
    sleep_df[sleep_df["Sex"] == "Both"]
    .groupby("Year")["Avg hrs per day sleeping"]
    .mean()
    .reset_index()
)
print(trend.to_string(index=False))
print("→ Columns used: Year + Avg hrs per day sleeping")

# 👥 ANALYSIS 2 — Group Comparison by Sex
print("\n\n👥 GROUP COMPARISON: Sleep by Sex")
print("-" * 45)
by_sex = (
    sleep_df.groupby("Sex")["Avg hrs per day sleeping"]
    .mean()
    .reset_index()
    .rename(columns={"Avg hrs per day sleeping": "Avg Sleep (hrs)"})
)
print(by_sex.to_string(index=False))
print("→ Columns used: Sex + Avg hrs per day sleeping")

# ANALYSIS 3 — Group Comparison by Age Group
print("\n\n GROUP COMPARISON: Sleep by Age Group")
print("-" * 45)
by_age = (
    sleep_df[sleep_df["Sex"] == "Both"]
    .groupby("Age Group")["Avg hrs per day sleeping"]
    .mean()
    .reset_index()
    .sort_values("Avg hrs per day sleeping", ascending=False)
)
print(by_age.to_string(index=False))
print("→ Columns used: Age Group + Avg hrs per day sleeping")

# ANALYSIS 4 — Weekday vs Weekend Comparison
print("\n\n COMPARISON: Weekday vs Weekend Sleep")
print("-" * 45)
by_daytype = (
    sleep_df[sleep_df["Sex"] == "Both"]
    .groupby("Type of Days")["Avg hrs per day sleeping"]
    .mean()
    .reset_index()
    .sort_values("Avg hrs per day sleeping", ascending=False)
)
print(by_daytype.to_string(index=False))
print("→ Columns used: Type of Days + Avg hrs per day sleeping")
```

# Exercise 7: Transformation Exercise
For each of the following unstructured data sources, propose a method to convert it into structured data. Explain your reasoning.

A series of blog posts about travel experiences.
Audio recordings of customer service calls.
Handwritten notes from a brainstorming session.
A video tutorial on cooking.

To effectively perform analysis, we must transform unstructured data into structured formats (rows and columns) that computers can query, sort, and model. Here are proposed methods for your specified sources:

1. Blog Posts about Travel Experiences
Method: Natural Language Processing (NLP) / Named Entity Recognition (NER).

Reasoning: Blog posts contain rich, subjective text. Using Python libraries like spaCy or NLTK, you can extract specific entities such as "Location," "Budget," "Rating," and "Activities." You can also perform Sentiment Analysis to assign a score (e.g., -1 to +1) to the traveler's experience. This converts narrative text into a structured dataframe where each row is a post and columns are extracted attributes.

2. Audio Recordings of Customer Service Calls
Method: Automatic Speech Recognition (ASR) + Text Mining.

Reasoning: Audio is inherently unstructured. First, you must use a transcription service (like OpenAI's Whisper or Google Cloud Speech-to-Text) to convert audio into raw text. Once transcribed, you can use Topic Modeling (such as Latent Dirichlet Allocation) to categorize the call (e.g., "Billing Issue," "Technical Support," "Cancellation"). This produces a structured table with columns like Call_Date, Topic_Category, Sentiment, and Resolution_Status.

3. Handwritten Notes from Brainstorming
Method: Optical Character Recognition (OCR) + Semantic Tagging.

Reasoning: The primary challenge here is digitizing the physical medium. Tools like Tesseract or Google Cloud Vision can convert images of handwriting into digital text. Once digitized, you can parse the text using regex (regular expressions) to identify key components like "Idea Title," "Action Items," or "Assigned Person," transforming a chaotic jumble of thoughts into a structured project management table.

4. Video Tutorial on Cooking
Method: Computer Vision (CV) + Metadata Extraction.

Reasoning: To structure video, you must synchronize visual data with timing. You can use Frame Extraction to identify specific ingredients shown on screen (using object detection models like YOLO). By pairing these identified objects with the video's timestamp and the audio transcript (the spoken instructions), you can create a structured "Recipe Recipe-Log" containing columns like Time_Marker, Action, Ingredients_Used, and Tools_Required

 # Exercise 9 : Export a dataframe to excel format and JSON format.
Create a simple dataframe.
Export the dataframe to an excel file.
Export the dataframe to a JSON file.

```python
import pandas as pd

# ── Step 1: Create a Simple DataFrame ─────────────────────
data = {
    "Name"      : ["Alice", "Bob", "Charlie", "Diana", "Eve"],
    "Age"       : [25, 30, 35, 28, 22],
    "City"      : ["New York", "Paris", "London", "Tokyo", "Sydney"],
    "Salary"    : [55000, 72000, 68000, 81000, 49000],
    "Department": ["HR", "Engineering", "Marketing", "Finance", "Design"]
}

df = pd.DataFrame(data)

print("=" * 50)
print("        CREATED DATAFRAME")
print("=" * 50)
print(df.to_string(index=False))
print(f"\n Shape: {df.shape[0]} rows × {df.shape[1]} columns")

# ── Step 2: Export to Excel ────────────────────────────────
excel_file = "employees.xlsx"
df.to_excel(excel_file, index=False, sheet_name="Employees")
print(f"\n Excel file saved  → {excel_file}")

# ── Step 3: Export to JSON ─────────────────────────────────
json_file = "employees.json"
df.to_json(json_file, orient="records", indent=4)
print(f" JSON file saved   → {json_file}")

# ── Step 4: Verify — Reload both files ────────────────────
print("\n\n VERIFICATION — Reloading saved files")
print("=" * 50)

# Reload Excel
df_excel = pd.read_excel(excel_file)
print(f"\n Reloaded from Excel ({excel_file}):")
print(df_excel.to_string(index=False))

# Reload JSON
df_json = pd.read_json(json_file)
print(f"\n Reloaded from JSON ({json_file}):")
print(df_json.to_string(index=False))
```
