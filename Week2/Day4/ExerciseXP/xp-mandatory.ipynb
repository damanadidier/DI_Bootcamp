Exercise 1: Understanding Data Visualization
Task: Explain why data visualization is important in data analysis.

Data visualization is crucial in data analysis because it allows analysts and stakeholders to quickly grasp complex data patterns, trends, and relationships. By representing data visually, it becomes easier to identify outliers, correlations, and insights that may not be immediately apparent in raw data. Visualizations can also enhance communication, making it easier to share findings with a wider audience, including those who may not have a technical background. Overall, data visualization helps to make informed decisions based on the data by providing a clear and intuitive way to interpret information.

Task: Describe the purpose of a line graph in data visualization.

A line graph is a type of data visualization that uses points connected by lines to represent the relationship between two variables over time or across categories. The primary purpose of a line graph is to show trends, patterns, and changes in data over a continuous range. It is particularly useful for tracking changes over time, such as sales growth, stock prices, or temperature variations. Line graphs can help identify upward or downward trends, seasonal patterns, and fluctuations in the data, making it easier for analysts to draw conclusions and make predictions based on the observed trends.

```python
# Exercise 2: Creating a Line Plot for Temperature Variation
# Objective: Create a simple line plot using Matplotlib that represents temperature variations over a week.
# Tasks:
# Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]).
# Label the x-axis as “Day” and the y-axis as “Temperature (°F)”.
# Add a title to the plot.
# Display the plot.

import matplotlib.pyplot as plt
# List of temperature values for each day of the week
temperatures = [72, 74, 76, 80, 82, 78, 75]
# Days of the week
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# Create a line plot
plt.plot(days, temperatures, marker='o')
# Label the axes
plt.xlabel('Day')
plt.ylabel('Temperature (°F)')
# Add a title to the plot
plt.title('Temperature Variation Over a Week')
# Display the plot
plt.grid()
plt.show()
```

```python
# Exercise 3: Visualizing Monthly Sales with a Bar Chart
# Objective: Generate a bar chart using Matplotlib to visualize monthly sales data for a retail store.
# Tasks:
# Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]).
# Label the x-axis as “Month” and the y-axis as “Sales Amount ($)”.
# Add a title to the bar chart.
# Display the plot.
import matplotlib.pyplot as plt
# List of sales values for each month
sales = [5000, 5500, 6200, 7000, 7500]
# Months of the year
months = ['January', 'February', 'March', 'April', 'May']
# Create a bar chart
plt.bar(months, sales, color='blue')
# Label the axes
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
# Add a title to the bar chart
plt.title('Monthly Sales Data for Retail Store')
# Display the plot
plt.grid(axis='y')
plt.show()
```

```python
# Exercise 4: Visualizing the Distribution of CGPA
# Objective: Create a histogram to visualize the distribution of students' CGPA. 
# Dataset Overview: Assume the CGPA data is categorized into ranges and loaded in a DataFrame named df.
# Tasks:
# Import necessary libraries.
# Use Seaborn's histplot to create a histogram of the CGPA categories.
# Customize the histogram with a specific color and add a title.
# Display the plot.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the dataset 
df = pd.read_csv('students.csv')
# Create a histogram of the CGPA categories (CGPA is stored as categorical ranges)
cgpa_col = 'What is your CGPA?'
order = sorted(df[cgpa_col].dropna().unique())
sns.histplot(data=df, x=cgpa_col, color='green', discrete=True)
plt.xticks(range(len(order)), order, rotation=45, ha='right')
# Add a title to the histogram
plt.title('Distribution of Students\' CGPA')
# Label the axes
plt.xlabel('CGPA')
plt.ylabel('Frequency')
# Display the plot
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python

# For exercises 5 use the same dataset students.csv

# Exercise 5: Comparing Anxiety Levels Across Different Genders
# Objective: Use a bar plot to compare the proportion of students experiencing anxiety across different genders.
# Dataset Overview: The dataset includes columns: ‘Do you have Anxiety?’ and ‘Choose your gender’.
# Tasks:
# Import necessary libraries.
# Use Seaborn to create a bar plot comparing anxiety levels across genders from the dataset df.
# Customize the plot with an appropriate color palette and add a title.
# Display the plot.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('students.csv')

# Convert 'Yes'/'No' to 1/0 so the bar height represents the proportion with anxiety
df['Anxiety'] = (df['Do you have Anxiety?'] == 'Yes').astype(int)

sns.barplot(
    data=df,
    x='Choose your gender',
    y='Anxiety',
    hue='Choose your gender',
    palette='Set2',
    legend=False,
    errorbar=None,
)
plt.title('Proportion of Students with Anxiety by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion with Anxiety')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Use the same dataset students.csv for exercise 6
# Exercise 6: Exploring the Relationship Between Age and Panic Attacks
# Objective: Create a scatter plot to explore the relationship between students’ age and the occurrence of panic attacks.
# Dataset Overview: The dataset includes columns: ‘Age’ and ‘Do you have Panic Attacks?’.
# Tasks:
# Import necessary libraries.
# Convert panic attack responses to numeric values (e.g., Yes=1, No=0).
# Use Seaborn’s scatterplot to create a scatter plot with ‘Age’ on the x-axis and numeric panic attack responses on the y-axis.
# Customize the plot to improve readability by adding labels, a title, and adjusting point styles.
# Display the plot.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('students.csv')
# Convert 'Yes'/'No' to 1/0 for panic attacks
df['Panic_Attack'] = (df['Do you have Panic attack?'] == 'Yes').astype(int)
# Create a scatter plot
sns.scatterplot(data=df, x='Age', y='Panic_Attack', alpha=0.6)
plt.title('Relationship Between Age and Panic Attacks')
plt.xlabel('Age')
plt.ylabel('Panic Attack (1=Yes, 0=No)')
plt.yticks([0, 1], ['No', 'Yes'])
plt.grid()
plt.tight_layout()
plt.show()
```
